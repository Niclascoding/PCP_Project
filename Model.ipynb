# Model Summary

```python
from neuralnet.ResNet import Connect4Model
from torchviz import make_dot
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
import random
from collections import deque
from torchinfo import summary
from IPython.display import Image
```

```python
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model = Connect4Model(device)

dummy_input = torch.randn(1, 3, 6, 7).to(device)
value, policy = model(dummy_input)

make_dot((value, policy), params=dict(model.named_parameters())).render("neuralnet/connect4_model_graph", format="png")
```
